# Question 2: What are the key assumptions of Simple Linear Regression?

> The key assumptions of Simple Linear Regression are the conditions that should be satisfied for the model to give reliable and accurate results. These assumptions are explained below:

Linearity
There should be a linear relationship between the independent variable (X) and the dependent variable (Y). This means that changes in X should result in proportional changes in Y.

Independence of Errors
The observations should be independent of each other. The error terms should not be related to one another.

Homoscedasticity
The variance of the error terms should be constant for all values of X. This means the spread of residuals remains the same across the regression line.

Normality of Errors
The error terms (residuals) should be normally distributed. This assumption is important for hypothesis testing and confidence intervals.

No Perfect Multicollinearity
Since simple linear regression has only one independent variable, this assumption is automatically satisfied. It means the independent variable should not be perfectly correlated with another variable.

These assumptions ensure that the regression model provides valid predictions and meaningful statistical inferences.

 # Question 3: What is heteroscedasticity, and why is it important to address in regression
models?
> Heteroscedasticity refers to a situation in a regression model where the variance of the error terms is not constant across all values of the independent variable.

In simple words, it means that the spread of residuals changes as the value of the independent variable changes. Instead of being evenly spread, the errors may increase or decrease, forming a cone or fan-shaped pattern.

Example:

While analyzing income and expenditure, lower-income individuals may show small variation in spending, whereas higher-income individuals may show much larger variation. This unequal spread of errors indicates heteroscedasticity.

Why is it important to address heteroscedasticity?

Incorrect Standard Errors
Heteroscedasticity leads to biased standard errors, which affects hypothesis testing.

Unreliable t-tests and F-tests
Because standard errors are incorrect, statistical tests may give misleading results.

Inefficient Estimates
Although regression coefficients may still be unbiased, they are not efficient, reducing the reliability of the model.

Poor Decision Making
In real-world data analysis, ignoring heteroscedasticity can lead to wrong conclusions and poor business decisions.

Therefore, detecting and correcting heteroscedasticity is important to ensure accurate and reliable regression results.

# Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
‚óè X = [1, 2, 3, 4, 5]
‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.
(Include your Python code and output in the code box below.)


```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])     # Dependent variable

# Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope (b): {slope}")
print(f"Intercept (a): {intercept}")

# Predict Y values using the model
Y_pred = model.predict(X)

# Plotting the data points and regression line
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

Explanation

LinearRegression() creates a simple linear regression model.

model.fit(X, Y) fits the model to your data.

model.coef_ and model.intercept_ give the slope and intercept of the regression line.

plt.scatter() plots the actual points, and plt.plot() plots the fitted regression line.

Expected Output:

A printed slope and intercept (like slope ‚âà 2, intercept ‚âà 0.1).

A graph showing blue dots (actual data) and a red straight line (regression line).



# Question 7: Fit a Multiple Linear Regression model on this sample data:
‚óè Area = [1200, 1500, 1800, 2000]
‚óè Rooms = [2, 3, 3, 4]
‚óè Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results.
(Include your Python code and output in the code box below.)

```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent variables (X) and dependent variable (Y)
X = df[['Area', 'Rooms']]
Y = df['Price']

# Fit Multiple Linear Regression
model = LinearRegression()
model.fit(X, Y)

# Print coefficients and intercept
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {list(zip(X.columns, model.coef_))}")

# Predict prices
Y_pred = model.predict(X)
print(f"Predicted Prices: {Y_pred}")

# -------------------------
# Check for multicollinearity using VIF
# Add constant for statsmodels
X_const = sm.add_constant(X)

# Calculate VIF
vif = pd.DataFrame()
vif['Features'] = X_const.columns
vif['VIF'] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
print("\nVariance Inflation Factor (VIF) Results:")
print(vif)
```

Question 8: Implement polynomial regression on the following data:
‚óè X = [1, 2, 3, 4, 5]

‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.
(Include your Python code and output in the code box below.)

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform features to polynomial (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the Polynomial Regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Predict values
Y_pred = model.predict(X_poly)

# Plot the original data points and polynomial curve
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Polynomial Regression Curve')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

# Question 9: Create a residuals plot for a regression model trained on this data:
‚óè X = [10, 20, 30, 40, 50]
‚óè Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.
(Include your Python code and output in the code box below.)


```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit Simple Linear Regression
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals Plot')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.show()

# Print residuals
print("Residuals:", residuals)
```

# Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

- As a data scientist predicting house prices using features like area, number of rooms, and location, encountering heteroscedasticity and multicollinearity can affect the accuracy and reliability of the regression model. Here‚Äôs how I would address these issues:

1. Addressing Heteroscedasticity

Heteroscedasticity occurs when the variance of residuals is not constant, which can lead to inefficient estimates. Steps to fix it:

Transform the dependent variable (Y):
Apply a logarithmic (log), square root, or Box-Cox transformation to stabilize variance.
Example: Price_log = log(Price)

Weighted Least Squares (WLS):
Give lower weight to observations with higher variance and higher weight to observations with lower variance.

Check residual plots again:
After applying transformations, plot residuals to ensure variance is more uniform.

2. Addressing Multicollinearity

Multicollinearity occurs when independent variables are highly correlated, which can make coefficient estimates unstable. Steps to fix it:

Calculate VIF (Variance Inflation Factor):
Identify which variables are causing multicollinearity (VIF > 10 is problematic).

Remove or combine correlated variables:
For example, if Area and Number of Rooms are highly correlated, you can remove one or combine them into a new feature like Room Density.

Use dimensionality reduction:
Techniques like Principal Component Analysis (PCA) can create uncorrelated components from correlated features.

3. Other Steps to Ensure a Robust Model

Feature scaling: Standardize variables to reduce bias due to scale differences.

Regularization methods: Use Ridge or Lasso regression to penalize large coefficients caused by multicollinearity.

Cross-validation: Split data into training and testing sets to ensure the model generalizes well.

Check residuals and model metrics: Ensure no patterns in residuals and evaluate metrics like RMSE and R¬≤.
